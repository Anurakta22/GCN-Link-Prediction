```python
import torch
import torch.nn.functional as F
import random

from torch_geometric.datasets import Planetoid
from torch_geometric.nn import GCNConv
from torch_geometric.transforms import RandomLinkSplit

from sklearn.metrics import roc_auc_score, average_precision_score

# Reproducibility
SEED = 42
torch.manual_seed(SEED)
random.seed(SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Hyperparameters
HIDDEN_DIM = 64
EMBED_DIM = 64
LR = 0.01
EPOCHS = 200
PATIENCE = 20
```

```python
dataset = Planetoid(root="data/Cora", name="Cora")
data = dataset[0]

print(data)
print("Nodes:", data.num_nodes)
print("Edges:", data.num_edges)
print("Node features:", dataset.num_node_features)
```

```python
transform = RandomLinkSplit(
    num_val=0.05,
    num_test=0.10,
    is_undirected=True,
    add_negative_train_samples=True
)

train_data, val_data, test_data = transform(data)

train_data = train_data.to(device)
val_data = val_data.to(device)
test_data = test_data.to(device)

print("Train edges:",
      (train_data.edge_label == 1).sum().item(),
      "+",
      (train_data.edge_label == 0).sum().item())

print("Val edges:",
      (val_data.edge_label == 1).sum().item(),
      "+",
      (val_data.edge_label == 0).sum().item())

print("Test edges:",
      (test_data.edge_label == 1).sum().item(),
      "+",
      (test_data.edge_label == 0).sum().item())
```

```python
class GCNEncoder(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels):
        super().__init__()
        self.conv1 = GCNConv(in_channels, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, out_channels)

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=0.5, training=self.training)
        x = self.conv2(x, edge_index)
        return x
```

```python
model = GCNEncoder(
    in_channels=dataset.num_node_features,
    hidden_channels=HIDDEN_DIM,
    out_channels=EMBED_DIM
).to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=LR)
```

```python
def decode(z, edge_label_index):
    return (z[edge_label_index[0]] * z[edge_label_index[1]]).sum(dim=1)

def compute_loss(z, edge_label_index, edge_label):
    pred = decode(z, edge_label_index)
    return F.binary_cross_entropy_with_logits(pred, edge_label)
```

```python
@torch.no_grad()
def evaluate(data):
    model.eval()
    z = model(data.x, data.edge_index)

    pred = decode(z, data.edge_label_index).cpu()
    label = data.edge_label.cpu()

    auc = roc_auc_score(label, pred)
    ap = average_precision_score(label, pred)

    return auc, ap
```

```python
best_val_auc = 0
patience_counter = 0
best_state = None

for epoch in range(1, EPOCHS + 1):
    model.train()
    optimizer.zero_grad()

    z = model(train_data.x, train_data.edge_index)

    loss = compute_loss(
        z,
        train_data.edge_label_index,
        train_data.edge_label.float()
    )

    loss.backward()
    optimizer.step()

    val_auc, val_ap = evaluate(val_data)

    if val_auc > best_val_auc:
        best_val_auc = val_auc
        patience_counter = 0
        best_state = model.state_dict()
    else:
        patience_counter += 1

    if epoch % 20 == 0:
        print(
            f"Epoch {epoch:03d} | "
            f"Loss {loss:.4f} | "
            f"Val AUC {val_auc:.4f} | "
            f"Val AP {val_ap:.4f}"
        )

    if patience_counter >= PATIENCE:
        print(f"\nEarly stopping at epoch {epoch}")
        break

model.load_state_dict(best_state)
```

```python
test_auc, test_ap = evaluate(test_data)

print("\nFINAL TEST RESULTS")
print("ROC-AUC:", round(test_auc, 4))
print("Average Precision:", round(test_ap, 4))
```

```python
@torch.no_grad()
def precision_at_k(data, k=100):
    model.eval()
    z = model(data.x, data.edge_index)

    probs = torch.sigmoid(decode(z, data.edge_label_index))
    labels = data.edge_label

    topk = probs.topk(k).indices
    precision = labels[topk].sum().item() / k

    return precision
```

```python
print("\nPrecision@K on Test Set")
for k in [20, 50, 100]:
    print(f"Precision@{k}: {precision_at_k(test_data, k):.4f}")
```

```python
@torch.no_grad()
def precision_recall(data, threshold=0.5):
    model.eval()
    z = model(data.x, data.edge_index)

    probs = torch.sigmoid(decode(z, data.edge_label_index))
    labels = data.edge_label

    preds = (probs >= threshold).float()

    tp = ((preds == 1) & (labels == 1)).sum().item()
    fp = ((preds == 1) & (labels == 0)).sum().item()
    fn = ((preds == 0) & (labels == 1)).sum().item()

    precision = tp / (tp + fp + 1e-8)
    recall = tp / (tp + fn + 1e-8)

    return precision, recall
```

```python
p, r = precision_recall(test_data)

print("Precision:", round(p, 4))
print("Recall:", round(r, 4))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

@torch.no_grad()
def plot_confusion_matrix(data, threshold=0.5):
    model.eval()
    z = model(data.x, data.edge_index)

    probs = torch.sigmoid(decode(z, data.edge_label_index)).cpu()
    preds = (probs >= threshold).int()
    labels = data.edge_label.cpu().int()

    cm = confusion_matrix(labels, preds)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                                  display_labels=["No Link", "Link"])

    disp.plot(cmap="Blues")
    plt.title("Confusion Matrix (Link Prediction)")
    plt.show()
```

```python
plot_confusion_matrix(test_data)
```

```python
from sklearn.metrics import roc_curve, auc

@torch.no_grad()
def plot_roc_curve(data):
    model.eval()
    z = model(data.x, data.edge_index)

    probs = torch.sigmoid(decode(z, data.edge_label_index)).cpu()
    labels = data.edge_label.cpu()

    fpr, tpr, _ = roc_curve(labels, probs)
    roc_auc = auc(fpr, tpr)

    plt.figure()
    plt.plot(fpr, tpr, label=f"ROC curve (AUC = {roc_auc:.4f})")
    plt.plot([0, 1], [0, 1], linestyle="--")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve")
    plt.legend(loc="lower right")
    plt.show()
```

```python
plot_roc_curve(test_data)
```

```python
from sklearn.metrics import precision_recall_curve

@torch.no_grad()
def plot_precision_recall_curve(data):
    model.eval()
    z = model(data.x, data.edge_index)

    probs = torch.sigmoid(decode(z, data.edge_label_index)).cpu()
    labels = data.edge_label.cpu()

    precision, recall, _ = precision_recall_curve(labels, probs)

    plt.figure()
    plt.plot(recall, precision)
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.title("Precision–Recall Curve")
    plt.show()
```

```python
plot_precision_recall_curve(test_data)
```

```python
import networkx as nx
import plotly.graph_objects as go
from torch_geometric.utils import to_networkx

@torch.no_grad()
def visualize_predicted_links(data, top_k=40):
    model.eval()
    z = model(data.x, data.edge_index)

    probs = torch.sigmoid(decode(z, data.edge_label_index))
    edge_pairs = data.edge_label_index.cpu().numpy()

    top_idx = probs.topk(top_k).indices.cpu().numpy()
    predicted_edges = edge_pairs[:, top_idx].T

    G = to_networkx(data, to_undirected=True)
    pos = nx.spring_layout(G, seed=42)

    # Existing edges
    edge_x, edge_y = [], []
    for u, v in G.edges():
        edge_x += [pos[u][0], pos[v][0], None]
        edge_y += [pos[u][1], pos[v][1], None]

    # Predicted edges
    pred_x, pred_y = [], []
    for u, v in predicted_edges:
        pred_x += [pos[u][0], pos[v][0], None]
        pred_y += [pos[u][1], pos[v][1], None]

    node_x = [pos[n][0] for n in G.nodes()]
    node_y = [pos[n][1] for n in G.nodes()]

    fig = go.Figure()

    fig.add_trace(go.Scatter(
        x=edge_x, y=edge_y,
        mode="lines",
        line=dict(color="lightgray"),
        name="Existing Links"
    ))

    fig.add_trace(go.Scatter(
        x=pred_x, y=pred_y,
        mode="lines",
        line=dict(color="red", width=2),
        name="Predicted Links"
    ))

    fig.add_trace(go.Scatter(
        x=node_x, y=node_y,
        mode="markers",
        marker=dict(size=6, color="blue"),
        name="Nodes"
    ))

    fig.update_layout(
        title="GCN Link Prediction (Red = Predicted Links)",
        width=900,
        height=700,
        showlegend=True
    )

    fig.show()
```

```python
visualize_predicted_links(test_data)
```
